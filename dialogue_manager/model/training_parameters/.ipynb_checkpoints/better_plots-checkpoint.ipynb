# Choosing Reward Function
After trying a lot to get the model to learn the policy, I changed my method a little bit. I did some runs with rule based listener for policy := (Select when P >= 0.5), 'always select' and 'always clarify' baselines to get an expression for the average rewards for each model. Using these expressions, I chose reward functions which set the average reward for P >= 0.95 highest.

# Model Choice for this run
This notebook showcases the policy learned by the listener model when the director simulation samples uses EM model to generate the descriptions and uses the following reward function:<br>
Reward_Success (Rs) = 0.95<br>
Reward_Failure (Rf) = -0.95<br>
Length Penalty Per Utterance (Rl) = -0.1<br>
<br>
One major change that I made to this simulation to achieve more 'human-like' behavior was to use sampled utterances for those cases when parser could not completely parse the sentence.

# Action Execution Description
### Speaker Simulation Actions
1- Speaker samples from the listener distribution to check whether the clarification question is about the target.<br>\
2- Rejection is just a no answer and Confirmation is a yes answer.
### Listener Model Actions
1- During selection, listener samples the color patch from posterior to determine which color patch to select.

```python
def get_policy(action_logs):
    incomplete_policy = dict()
    complete_policy = dict()
    for item in action_logs:
#         print(item)
        dist, act, prev_act, conv_len, incomplete = item
        dist = np.array(dist)
        if incomplete:
            to_fill = incomplete_policy
        else:
            to_fill = complete_policy
            
        dist.sort()
#         max_dist = dist[2] - dist[1]
        max_dist = dist[2]

        temp = max_dist * 10
        bucket = math.floor(temp) + 0.5 if temp >= (math.floor(temp) + 0.5) else math.floor(temp)
        bucket /= 10
        to_fill.setdefault(conv_len, dict())
        to_fill[conv_len].setdefault(bucket, dict())
        to_fill[conv_len][bucket].setdefault(act, 0)
        to_fill[conv_len][bucket][act] += 1
    return incomplete_policy, complete_policy
```

```python
def print_policy(policy_dict, \
    action_names=['NONE', 'IDENTIFY', 'A?', 'CONFIRMATION', 'SELECT', 'REJECTION', 'A or B?', 'A or B or C?']):
    for length in policy_dict:
        print('Conversation Turn:', length + 1)
        for prob_bucket in policy_dict[length]:
            print('\t Max Prob >= %f and <= %f' % (prob_bucket, prob_bucket + 0.05))
            for action in policy_dict[length][prob_bucket]:
                action_name = action_names[action]
                count = policy_dict[length][prob_bucket][action]
                print('\t\t%s Count:' % action_name, count)
```

# Learned Weights
This the analysis of learned weights for actions: 'ASK_CLARIFICATION' and 'SELECT'. The results are displayed in the following format: <br>
FEATURE_NAME: [SampleSimulationModelFeatureCoeff, EMModelSimulationModelFeatureCoeff]

```python
#analyzing learned weights
# actions = ['NONE', 'IDENTIFY', 'ASK_CLARIFICATION', 'CONFIRMATION', 'SELECT', 'REJECTION']
def print_learned_params(weight_dict, top_features, bottom_features, num_models, model_ix, \
    actions=['NONE', 'IDENTIFY', 'ASK_CLARIFICATION', 'CONFIRMATION', 'SELECT', 'REJECTION']):
    for key in weight_dict:
        if 'weight' in key:
            weight_shape = weight_dict[key].shape
            weights = weight_dict[key]
            weight_features = []
            for i in range(weight_shape[0]):
                top_features.setdefault(actions[i], dict())
                bottom_features.setdefault(actions[i], dict())
                
                sorted_weights = weights[i].sort()
                for j in range(10):
                    feature_coeff = float(sorted_weights.values[-j - 1])
                    if sorted_weights.indices[-j - 1] < 3:
                        feature_name = 'Patch %d Prob' % (sorted_weights.indices[-j - 1],)
                    else:
                        action_ix = (sorted_weights.indices[-j - 1] - 3) % len(actions)
                        turn = int(((sorted_weights.indices[-j - 1] - 3) // len(actions)) + 1)
                        feature_name = actions[action_ix] + ' action at turn ' + str(turn) + \
                            ' with speaker: %s' % ('director' if turn % 2 else 'matcher',)
                        
                    top_features[actions[i]].setdefault(feature_name, [-1 for i in range(num_models)])
                    top_features[actions[i]][feature_name][model_ix] = round(feature_coeff, 3)
                    
                for j in range(10):
                    feature_coeff = float(sorted_weights.values[j])
                    if sorted_weights.indices[j] < 3:
                        feature_name = 'Patch %d Prob' % (sorted_weights.indices[j],)
                    else:
                        action_ix = (sorted_weights.indices[j] - 3) % len(actions)
                        turn = ((sorted_weights.indices[j] - 3) // len(actions)) + 1
                        feature_name = actions[action_ix] + ' action at turn ' + str(turn) + \
                            ' with speaker: %s' % ('director' if turn % 2 else 'matcher',)
                        
                    bottom_features[actions[i]].setdefault(feature_name, [-1 for i in range(num_models)])
                    bottom_features[actions[i]][feature_name][model_ix] = round(feature_coeff, 3)
    return top_features, bottom_features

top_features = dict()
bottom_features = dict()

weight_dicts = [torch.load('../weights/trained_modelLIST_SAMPLESAMPLEDCOMPOSITE0.0.pt')]

for i, wt_dt in enumerate(weight_dicts):
#     print(wt_dt)
    top_features, bottom_features = print_learned_params(wt_dt, top_features, \
                                        bottom_features, len(weight_dicts), i)

print('Displaying Top 10 Features. \n$$$$$$$$$$$$$$$$$$$$$$$$$$$$')
for action in top_features.keys():
    print('Feature Coefficients for %s' % (action,))
    for k, v in top_features[action].items():
        print('\t%s: %s' % (k, v))
    print('********************************')
print('Displaying Bottom Features. \n$$$$$$$$$$$$$$$$$$$$$$$$$$$$')
for action in bottom_features.keys():
    print('Feature Coefficients for %s' % (action,))
    for k, v in bottom_features[action].items():
        print('\t%s: %s' % (k, v))
    print('********************************')
```

```python
def conduct_test_phase(models, cic, g, t, n, fs, irmap, rimap):
#     for i in range(0, int(os.getenv('TEST_CONVERSATIONS'))):
#     mis_parsed = 0
    max_counts = [150, 150, 300]
    counts = [0, 0, 0]
    not_completed = []
    for i in range(0, 1000):
        item = cic.get_index_item(i)
        rix = item['row_index']
        if cic._df.iloc[rix]['condition'] == 'far':
            if counts[0] >= max_counts[0]:
                continue
            counts[0] += 1
        elif cic._df.iloc[rix]['condition'] == 'split':
            if counts[1] >= max_counts[1]:
                continue
            counts[1] += 1
        elif cic._df.iloc[rix]['condition'] == 'close':
            if counts[2] >= max_counts[2]:
                continue
            counts[2] += 1
        else:
            print('Error', cic._df.iloc[rix]['condition'])
        if (os.getenv('SIMULATOR') == 'SAMPLE') or (i in cic._parse_mistakes):
            models[0] = SQAModel(num_actions, cic=cic, ix=i)
        elif os.getenv('SIMULATOR') == 'MODEL':
            models[0] = QAModel(num_actions)

        colors = item['rgb_colors']
        clr_trans = item['x_colors']

        shuffled_colors = [tup for tup in zip(clr_trans, colors)]
        np.random.shuffle(shuffled_colors)
        clr_pair = {'S':clr_trans, 'L': [shuffled_colors[i][0] for i in range(len(shuffled_colors))], \
            'S_RGB': colors, 'L_RGB' : [shuffled_colors[i][1] for i in range(len(shuffled_colors))]}

        dman = manager(clr_pair, *models, (g, t, n, fs), \
            (irmap, rimap), cic=cic, batch_dict=item)
        speaker = 'S'
        u, lf, done = dman.get_next_move()
#         print(speaker + ':', lf)
        while not done:
            speaker = 'L' if speaker == 'S' else 'S'
#             print('Vector State:', dman._state_rep.get_vector_representation())
            u, lf, done = dman.get_next_move()
            if done and not 'SELECT' in lf:
                not_completed.append(i)
        
        if counts[0] >= max_counts[0] and counts[1] >= max_counts[1] and \
            counts[2] >= max_counts[2]:
            
            print('Finished')
            return models[1], not_completed
#             print(speaker + ':', lf)
#         print('################################')
#     print('misparsed_count:', mis_parsed)
    return models[1], not_completed
```

```python
def get_move_percentage(action_choices):
    action_map = {2 : 0, 6 : 1, 7 : 2, 4 : 3}
    count = np.array([0, 0, 0, 0])
    total_count = 0
#     prev = 4
    for _, action, pa, le, ic in action_choices:
#         if prev == 4:
        count[action_map[action]] += 1
        total_count += 1
#         prev = action
    return (count/total_count)[:3]
```
